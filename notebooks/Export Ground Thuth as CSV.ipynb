# Dependencies

```python
import pandas as pd
from glob import glob

pd.options.display.max_columns = 100
pd.options.display.max_colwidth = 200
```

```python
FOLDER_APPS = '../../icao-reqs-pybossa/pybossa_apps'
FOLDER_CSVS = 'data/recover_taskruns'

OUTPUT_FILE = '../data/pybossa/volunt_and_experts_r1_extra_corr.csv'
```

# Helper Methods 

```python
def load_list_of_csv_as_dataframe(list_csvs):
    df = pd.DataFrame()
    contents = [pd.read_csv(csv_file) for csv_file in list_csvs]
    return df.append(contents)
```

# Load Data 

### icaoreqs_valid_volunt_app

```python
list_csvs = glob(f'{FOLDER_APPS}/icaoreqs_valid_volunt_app/{FOLDER_CSVS}/*.csv')

print(list_csvs)
```

```python
df_volunteers = load_list_of_csv_as_dataframe(list_csvs)
df_volunteers = df_volunteers[df_volunteers.user_id.isin([3, 5, 6])]

print(df_volunteers.shape)
df_volunteers.head()
```

### icaoreqs_valid_expert_app_r1

```python
list_csvs = glob(f'{FOLDER_APPS}/icaoreqs_valid_expert_app_r1/{FOLDER_CSVS}/*.csv')

print(list_csvs)
```

```python
df_experts_r1 = load_list_of_csv_as_dataframe(list_csvs)
    
print(df_experts_r1.shape)
df_experts_r1.head()
```

### icaoreqs_experts_extra_app

```python
list_csvs = glob(f'{FOLDER_APPS}/icaoreqs_experts_extra_app/{FOLDER_CSVS}/*.csv')

print(list_csvs)
```

```python
df_experts_extra = load_list_of_csv_as_dataframe(list_csvs)

print(df_experts_extra.shape)
df_experts_extra.head()
```

### icaoreqs_valid_expert_app_r1_corr

```python
list_csvs = glob(f'{FOLDER_APPS}/icaoreqs_valid_expert_app_r1_corr/{FOLDER_CSVS}/*.csv')

print(list_csvs)
```

```python
df_experts_corr = load_list_of_csv_as_dataframe(list_csvs)

print(df_experts_corr.shape)
df_experts_corr.head()
```

# Append All 

```python
df = pd.concat([df_volunteers, df_experts_r1, df_experts_extra, df_experts_corr])
df.finish_time = pd.to_datetime(df.finish_time)

print(df.shape)
df.head()
```

# Sanity Checks 

```python
df_votes = df.sort_values(by=['img_path', 'user_id', 'project_id', 'finish_time'])
df_votes.drop_duplicates(subset=['img_path', 'user_id'], keep='last', inplace=True)

df_votes['votes'] = df_votes.groupby('img_path')['user_id'].transform('count')
```

```python
df_errors = df_votes[df_votes.votes != 5]

print(df_errors.shape)
df_errors
```

```python
df_final = df_votes[df_votes.votes == 5]

assert(df_final.user_id.nunique() == 5)
assert(df_final.img_path.nunique() == 5763)
assert(df_final.groupby('img_path')['user_id'].count().max() == 5)
assert(df_final.groupby('img_path')['user_id'].count().min() == 5)
```

# Save

```python
df_final.to_csv(OUTPUT_FILE, index=False)
```
